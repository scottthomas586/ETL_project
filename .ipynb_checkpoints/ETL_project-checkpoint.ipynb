```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#store stock CSV data into DataFrame
all_stocks = "Resources/all_stocks_5yr.csv"
all_stocks_df = pd.read_csv(all_stocks)
all_stocks_df
```

```python
#store ticker symbol CSV data into DataFrame
ticker = "Resources/symbols_valid_meta.csv"
ticker_df = pd.read_csv(ticker)
ticker_df
```

```python
#create new data with select columns
all_stocks_trim = all_stocks_df[['date', 'open', 'close', 'volume', 'Name']].copy()
all_stocks_trim
```

```python
#create new data with select columns
ticker_trim = ticker_df[['Symbol', 'Security Name', 'Listing Exchange']].copy()
ticker_trim
```

```python
#transform stock data DataFrame
clean_all_stocks = all_stocks_trim.rename(columns={'Name': 'ticker_symbol'})
clean_all_stocks
```

```python
#transform ticker data DataFrame
clean_ticker = ticker_trim.rename(columns={'Symbol': 'ticker_symbol',
                                          'Security Name': 'company',
                                          'Listing Exchange': 'exchange'})
clean_ticker.set_index('ticker_symbol', inplace=True)
clean_ticker
```

```python
clean_ticker['exchange'].value_counts()
```

```python
clean_ticker_names = clean_ticker.replace({'Q': 'NASDAQ',
                                          'N': 'NYSE',
                                          'P': 'NYSEARCA',
                                          'Z': 'BATS',
                                          'A': 'NYSEAMERICAN'})
clean_ticker_names
```

```python
#create database connection
username = 'postgres'
password = 'V3ntur3s'
host = '127.0.0.1'
port = '5432'
database = 'stock_data'

db_url = 'postgresql://{}:{}@{}:{}/{}'.format(username, password, host, port, database)
```

```python
engine = create_engine(db_url)
conn = engine.connect()
```

```python
engine.table_names()
```

```python
#use pandas to load CSV converted DataFrame into database
clean_all_stocks.to_sql(name='stock_data', con=engine, if_exists='append', index=False)
```

```python
clean_ticker_names.to_sql(name='ticker_data', con=engine, if_exists='append', index=True)
```
